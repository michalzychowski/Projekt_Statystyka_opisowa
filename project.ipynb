### 3. Utwórz szereg rozdzielczy o szerokości klas h.                       
Szerokość klas h = R/k, gdzie R to rozstęp wartość (max-min).           
Liczbę klas k wyznaczamy ze wzorów pomocniczych, $k\leq 5\ln(n)$ lub $k = 1 + 3,322\ln(n)$ lub $k=\sqrt{n}$. 



```r
h <- rozstep/70
k <- 70
szereg <- cut(Pomiary$Pomiar, breaks = seq(min(Pomiary$Pomiar), max(Pomiary$Pomiar), by = (max(Pomiary$Pomiar) - min(Pomiary$Pomiar))/k))
szereg
```

### 4. Wyznacz liczności klas szeregu rozdzielczego.

```r
licznosci <- table(szereg)
licznosci
```

### 5. Wykonaj wykres paskowy liczności pomiarów i histogram szeregu rozdzielczego.

```r
stripchart(licznosci, method = "stack", main='Liczności pomiarów', vertical = TRUE, pch=21, col = "blue")
hist(Pomiary$Pomiar, breaks = k, col = "light blue", main = "Histogram")
```
